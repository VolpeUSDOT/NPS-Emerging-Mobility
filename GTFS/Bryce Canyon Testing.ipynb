```python
import requests
import re
import json

# Function to fetch vehicle points
def get_vehicle_points():
    url = 'https://brycecanyonshuttle.com/Services/JSONPRelay.svc/GetMapVehiclePoints?apiKey=8882812681&isPublicMap=true'
    response = requests.get(url)
    jsonp_data = response.text
    json_data = re.sub(r'^[^(]*\((.*)\)$', r'\1', jsonp_data)
    return json.loads(json_data)

# Function to fetch description for a specific vehicle ID
def get_vehicle_description(vehicle_id):
    url = f'https://brycecanyonshuttle.com/Services/JSONPRelay.svc/GetVehicleRouteStopEstimates?quantity=1&vehicleIdStrings={vehicle_id}'
    response = requests.get(url)
    jsonp_data = response.text
    json_data = re.sub(r'^[^(]*\((.*)\)$', r'\1', jsonp_data)
    
    # Parse the JSON data
    data = json.loads(json_data)
    
    # Check if the data contains estimates and extract the first description
    if isinstance(data, list) and len(data) > 0:
        estimates = data[0].get('Estimates', [])
        if len(estimates) > 0:
            description = estimates[0].get('Description', 'No Description available')
            return description
    return "No Description available"

# Fetch vehicle points and extract all vehicle IDs
vehicle_points = get_vehicle_points()

# Extract vehicle IDs
vehicle_ids = [vehicle['VehicleID'] for vehicle in vehicle_points]

# Fetch and print description for each vehicle ID
for vehicle_id in vehicle_ids:
    description = get_vehicle_description(vehicle_id)
    print(f"Vehicle ID: {vehicle_id}, Description: {description}")
```

```python
import requests
import re
import json
import time

# Function to fetch vehicle points and return the active VehicleID with Latitude and Longitude
def fetch_vehicle_locations():
    url = 'https://brycecanyonshuttle.com/Services/JSONPRelay.svc/GetMapVehiclePoints?apiKey=8882812681&isPublicMap=true'
    
    # Fetch data from the URL
    response = requests.get(url)
    jsonp_data = response.text
    
    # Clean the JSONP format to extract JSON
    json_data = re.sub(r'^[^(]*\((.*)\)$', r'\1', jsonp_data)
    
    # Parse the JSON data
    vehicle_data = json.loads(json_data)
    
    # Iterate over each vehicle point and print VehicleID, Latitude, and Longitude
    print("\nActive Vehicles:")
    for vehicle in vehicle_data:
        vehicle_id = vehicle.get("VehicleID", "Unknown")
        latitude = vehicle.get("Latitude", "Unknown")
        longitude = vehicle.get("Longitude", "Unknown")
        print(f"Vehicle ID: {vehicle_id}, Latitude: {latitude}, Longitude: {longitude}")

# Run this in a loop to update every 30 seconds
try:
    while True:
        fetch_vehicle_locations()
        # Wait for 30 seconds before the next update
        time.sleep(30)
except KeyboardInterrupt:
    print("Stopped updating vehicle locations.")
```
