```python
path = "Park Unit Scraping Information.csv"
parks = pd.read_csv(path, encoding='latin-1')
parks.head()
```

```python
## For context, here is the alerts site for each national park
alert_title_list = []
alerts_description_list = []
alerts_full_description_list = []
alert_title_counter_list = []
alerts_description_counter_list = []

for x in parks['index site']:
    
    try:
        #open driver
        driver = webdriver.Chrome(service=service, options=options)
        
        #pull alerts & conditions page
        driver.get(x)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'lxml')

        alert_title_counter = 0
        alert_title = []
        for y in soup.find_all("h3"):
            if "Alerts-severity" in str(y):
                alert_title.append(y.text)
                alert_title_counter +=1
                

        alerts_description_counter = 0
        alerts_description = []
        alerts_full_description = []
        for z in soup.find_all("h4"):
            if "Alert-title" in str(z):
                alerts_description.append(z.text)
                alerts_full_description.append(z.find_next_sibling("span").text)
                alerts_description_counter +=1

        
        print(x, alert_title_counter, alerts_description_counter)
        alert_title_list.append(alert_title)
        alerts_description_list.append(alerts_description)
        alerts_full_description_list.append(alerts_full_description)
        alert_title_counter_list.append(alert_title_counter)
        alerts_description_counter_list.append(alerts_description_counter)
        driver.close()
    except:
        print(x, "didnt work")
        driver.close()
        pass
```

```python
len(alert_title_list), len(alerts_description_list), len(alerts_full_description_list), len(alert_title_counter_list), len(alerts_description_counter_list)
```

```python
parks['Alert title list']=alert_title_list
parks['Alert description']=alerts_description_list
parks['Alert full description']=alerts_full_description_list
parks['Number of alerts']=alert_title_counter_list
parks['Number of alerts with description']=alerts_description_counter_list
```

```python
parks.columns
```

```python
road_closed_list = []
for x in range(0,len(parks)):
    road_closed = 0
    
    for y in parks['Alert description'][x]:
        if "lane closure" in str(y).lower():
            road_closed = 1
        if "road closure" in str(y).lower():
            road_closed = 1
        if "road closed" in str(y).lower():
            road_closed = 1
        if "road is closed" in str(y).lower():
            road_closed = 1
        if ("road" in str(y).lower())&("closed" in str(y).lower()):
            road_closed=1
    for y in parks['Alert full description'][x]:
        if "lane closure" in str(y).lower():
            road_closed = 1
        if "road closure" in str(y).lower():
            road_closed = 1
        if "road closed" in str(y).lower():
            road_closed = 1
        if "road is closed" in str(y).lower():
            road_closed = 1
        if ("road" in str(y).lower())&("closed" in str(y).lower()):
            road_closed=1
        
        
    road_closed_list.append(road_closed)
    if road_closed==1:
        print(parks['Alpha'][x])

parks['Road Closure Alert Flag']=road_closed_list
print(sum(parks['Road Closure Alert Flag']))
```
