```python
import pandas as pd
import openai
import numpy as np
import re
```

```python
with open("C:/Users/Drew.Quinton/Documents/GitHub/NPS-Emerging-Mobility/ARTNAPI.txt", "r", encoding="utf-8") as f:
    openai_api_key = f.read()

client = openai.OpenAI(
    api_key = openai_api_key,
    base_url="http://10.75.42.137:4000/" 
    )
```

```python
user_prompt = 'tell me if the following is a description of a road closure. If so, respond with "yes", if not, respond with "no" : '
system_prompt = 'You will be given a series of phrases. These phrases are general alerts to users visiting national parks. Your job is to identify, alerts that indicate that a road, lane, or parking lot is closed or reduced in capacity. Do not identify trails that are closed or reduced in capacity.'

alerts_df = pd.read_csv('alerts_api_tracker.csv')
# alerts_df.drop_duplicates(inplace=True)
alerts_df['Road Closure'] = np.nan

total_alerts = len(alerts_df)
alerts_processed = 0
formatted_closure_response = ""

print(f'processing {total_alerts} alerts...')

for index, row in alerts_df.iterrows():
    
    description = str(row['description'])
    
    if pd.isna(row['Road Closure']):
        response = client.chat.completions.create(
                            
            model = "GPT-4.1-mini", # model to send to the proxy. One of: GPT-4o-Mini, GPT-4o, o1-Mini, o4-Mini, GPT-4.1, GPT-4.1-mini, GPT-4.1-nano,
            # text-embedding-3-small, text-embedding-3-large, MAI-DS-R1, Llama-3.2-Vision, GPT-5, GPT-5-Mini, GPT-5-Nano , gpt-4o-transcribe
            messages = [
                {
                    "role": "system",
                    "content": system_prompt
                },
                {
                    "role": "user",
                    "content": user_prompt + description
                }
            ]
        )
    
        closure_response = response.choices[0].message.content
        formatted_closure_response = re.sub(r'[^A-Za-z0-9\s]', '', closure_response.lower())
        alerts_df.at[index, 'Road Closure'] = formatted_closure_response
    
    if alerts_processed % 100 == 0:
        alerts_df.to_csv('All_Alerts_with_Road_Closure_AI_Flag.csv')
        print(f'saved All_Alerts_with_Road_Closure_AI_Flag.csv')
    
    percent_complete = round(((alerts_processed/total_alerts)*100), 1)
    print(f'{formatted_closure_response} | finished line {alerts_processed}    |   {percent_complete}% complete')
    alerts_processed = alerts_processed +1
    
alerts_df.to_csv('All_Alerts_with_Road_Closure_AI_Flag.csv')
print(f'saved All_Alerts_with_Road_Closure_AI_Flag.csv')
```
