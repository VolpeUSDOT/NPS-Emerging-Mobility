```python
import requests
import json 
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import datetime

```

```python
res = requests.get('https://maps2.dcgis.dc.gov/dcgis/rest/services/DCGIS_DATA/Transportation_Bikes_Trails_WebMercator/MapServer/5/query?where=1%3D1&outFields=*&outSR=4326&f=json')
```

```python
data = res.text
df = json.loads(data) # json.dumps take a dictionary as input and returns a string as output.

#df = json.dumps(json.loads(data)) # json.dumps take a dictionary as input and returns a string as output.
#df= # json.loads take a string as input and returns a dictionary as output.
```

```python
tidal_basin_stations = [
    'Jefferson Memorial',
    'Hains Point/Buckeye & Ohio Dr SW',
    'Ohio Dr & West Basin Dr SW / MLK & FDR Memorials',
    '17th St & Independence Ave SW',
    'Lincoln Memorial',
    'Maine Ave & 9th St SW',
    'Jefferson Dr & 14th St SW',
    'USDA / 12th & C St SW',
    "10th St & L'Enfant Plaza SW",
    "Maine Ave & 7th St SW",
    'Henry Bacon Dr & Lincoln Memorial Circle NW'
]
```

```python
station_id_list = []
station_name_list = []
capacity_list = []
num_docks_available_list = []
num_bikes_available_list = []
num_ebikes_available_list = []
lat_list = []
lon_list = []


for x in df['features']:
    print(x)
    if x['attributes']['NAME'] in tidal_basin_stations:
        station_id_list.append(x['attributes']['STATION_ID'])
        station_name_list.append(x['attributes']['NAME'])
        capacity_list.append(x['attributes']['CAPACITY'])
        num_docks_available_list.append(x['attributes']['NUM_DOCKS_AVAILABLE'])
        num_bikes_available_list.append(x['attributes']['NUM_BIKES_AVAILABLE'])
        num_ebikes_available_list.append(x['attributes']['NUM_EBIKES_AVAILABLE'])
        lat_list.append(x['attributes']['LATITUDE'])
        lon_list.append(x['attributes']['LONGITUDE'])

      #  print(y['attributes']['NAME'])
#    break
```

```python
df_update = pd.DataFrame({
    "station name":station_name_list,
    "station id":station_id_list,
    "capacity": capacity_list,
    "docks available": num_docks_available_list,
    "bikes available": num_bikes_available_list,
    "ebikes available": num_ebikes_available_list,
    "latitude": lat_list,
    "longitude": lon_list,
#    "event end date": end_date_list
})
```

```python
now = datetime.datetime.now()
year = '{:02d}'.format(now.year)
month = '{:02d}'.format(now.month)
day = '{:02d}'.format(now.day)
hour = '{:02d}'.format(now.hour)
minute = '{:02d}'.format(now.minute)
day_month_year = '{}-{}-{}'.format(year, month, day)

df_update['date'] = day_month_year
df_update['hour'] = hour
```

```python
df_update.to_excel("CaBi-Cherry Blossoms Data.xlsx",index=False)
```
