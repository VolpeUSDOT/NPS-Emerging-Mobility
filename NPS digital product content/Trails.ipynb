```python
# import dataset
path = "park_scrape_content_dataset.xlsx"
park_scrape_dataset2 = pd.read_excel(path)
park_scrape_dataset2.head
```

```python
## convert all text to lowercase to avoid case sensitivity issues
park_scrape_dataset2['content'] = park_scrape_dataset2['content'].str.lower()
park_scrape_dataset2.head
```

```python
pass_words =  [
       "timed entry", "reservation", "timed-entry", "managed access", "timed access", "timed-access", ]
```

```python
def count_ped_words(text):
    word_counts = {word: text.lower().count(word) for word in ped_words}
    return word_counts
word_counts_df = park_scrape_dataset2['content'].apply(count_ped_words).apply(pd.Series)
word_counts_df['total'] = word_counts_df.sum(axis=1)
word_counts_df.head()
```

```python
park_scrape_dataset2 = pd.concat([park_scrape_dataset2, word_counts_df], axis=1)
park_scrape_dataset2 = park_scrape_dataset2.drop("content", axis="columns")
park_scrape_dataset2.head
```

```python
total_parks_counts = clusterparks.groupby('Label').size().reset_index(name='Total_Parks')
total_parks_counts['Total_Parks'] = total_parks_counts['Total_Parks'].astype(int) 
filtered_df = clusterparks[clusterparks['total'] > 1]
ped_parks_counts = filtered_df.groupby('Label').size().reset_index(name='ped_Parks')
label_counts = pd.merge(total_parks_counts, ped_parks_counts, on='Label', how='left')
label_counts['ped_Parks'].fillna(0, inplace=True)
label_counts['Percent_ped'] = round((label_counts['ped_Parks'] / label_counts['Total_Parks']) * 100, 1)
label_counts['ped_Parks'] = label_counts['ped_Parks'].astype(int)
label_counts.head()
```

```python
total_row = label_counts.sum(numeric_only=True)
total_row['Label'] = 'Total'
total_counts = label_counts.append(total_row, ignore_index=True)
total_counts['ped_Parks'] = total_counts['ped_Parks'].astype(int)
total_counts['Total_Parks'] = total_counts['Total_Parks'].astype(int)
total_counts.loc[total_counts['Label'] == 'Total', 'Percent_ped'] = round((total_counts.loc[total_counts['Label'] == 'Total', 'ped_Parks'] / total_counts.loc[total_counts['Label'] == 'Total', 'Total_Parks']) * 100, 1)
total_counts.tail()
```

```python
path = "pyvpages.xlsx"
sites = pd.read_excel(path)
sites.head
```

```python
website_stems = sites[['park', 'url']]
website_stems.head
```

```python
def extract_text(url):
    start = url.find('planyourvisit/') + len('planyourvisit/')
    end = url.find('.htm', start)
    if start != -1 and end != -1:
        return url[start:end]
    return None

sites['url_stem'] = sites['url'].apply(extract_text)
sites.head
```

```python
# Define parking words
ped_words =  [
       "bike", "cycl", "pedestrian", "on foot", "on-foot", "walking"]

# Function to count accessibility words in text
def count_ped_words(text):
    word_counts = {word: text.lower().count(word) for word in ped_words}
    return word_counts

# Apply the function to 'url_stem' column and convert to DataFrame
word_counts_df = sites['url_stem'].apply(count_ped_words).apply(pd.Series)
word_counts_df['ped_page'] = word_counts_df.apply(lambda row: 1 if row.any() >= 1 else 0, axis=1)
word_counts_df = word_counts_df[['ped_page']]

# Display the resulting DataFrame
print(word_counts_df.head())
sites = pd.concat([sites, word_counts_df], axis=1)
sites.to_csv("pedpagelist.csv")
sites.head
```

```python
park_word_counts = sites.groupby('park')['ped_page'].sum().reset_index()
# Display the resulting DataFrame
park_word_counts.rename(columns={'park':"Park Alpha"}, inplace = True)
park_word_counts['Park Alpha'] = park_word_counts['Park Alpha'].str.upper()
pagesbycluster = pd.merge(park_word_counts, clusters, on="Park Alpha", how="left")
print(pagesbycluster)
```

```python
filtered_df = pagesbycluster[pagesbycluster['ped_page'] >= 1]
print(filtered_df)
```

```python
ev_parks_counts = filtered_df.groupby('Label').size().reset_index(name='ped_Parks')
label_counts = pd.merge(total_parks_counts, ev_parks_counts, on='Label', how='left')
label_counts['ped_Parks'].fillna(0, inplace=True)
label_counts['Percent_ped'] = round((label_counts['ped_Parks'] / label_counts['Total_Parks']) * 100, 1)
label_counts['ped_Parks'] = label_counts['ped_Parks'].astype(int)
label_counts.head()
```

```python
plt.figure(figsize=(10, 6)) 
bars = plt.bar(label_counts['Label'], label_counts['Percent_ped'], color='#C56C39')

for i, bar in enumerate(bars):
    ev_parks = label_counts['ped_Parks'].iloc[i]
    total_parks = label_counts['Total_Parks'].iloc[i]
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.25,
             f"{ev_parks}/{total_parks}",
             ha='center', va='bottom')

plt.xlabel('Cluster')
plt.ylabel('Percentage of Parks with Dedicated Bike/Pedestrian Pages')
plt.title('Percentage of Parks with Dedicated Bike/Pedestrian Pages by Cluster')
plt.xticks(rotation=45, ha='right') 
plt.ylim(0, 50)
plt.tight_layout()
plt.savefig("PedPagesUpdated.png")
plt.show()
```
